### *`DICTINOARY`*

- Dictionary is unordered(Before 3.6) and ordered(after 3.6).
- Dictionaries are a collection in python which consists of key-value pairs
- Dictionaries are indexable and mutable
- Dictionary items are key-value pairs.
- Items in a dictionary separated by commas.
- keys and values are separated by colon
- dictionaries dont allow duplicate keys.
- Empty dictionary can be created using {}.

A Python dictionary is a data structure that stores key/value pairs. Each element of a Python dictionary is a key/value pair. The keys are like an index operator. In a list, the indexes are integers. In a dictionary, the key must be a hashable object. A dictionary cannot contain duplicate keys. Each key maps to one value. A key and its corresponding value form an item (or entry) stored in a dictionary. The data structure is a called a “dictionary” because it resembles a word dictionary, where the words are the keys and the words’ definitions are the values. A dictionary is also known as a map, which maps each key to a value. It enables fast retrieval, deletion, and updating of the value by using the key.
